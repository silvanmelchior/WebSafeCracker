# Imports

```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

# Helpers

```python
def store_graph(functions, names, filename, xmin=-1, xmax=1, ymin=-1, ymax=1, points=1000):
    dump = {
        'signals': {},
        'limits': {
            'xmin': float(xmin),
            'xmax': float(xmax),
            'ymin': float(ymin),
            'ymax': float(ymax)
        }
    }
    
    plt.figure(figsize=(10, 6))
    x_vals = np.linspace(xmin, xmax, points, dtype=float)
    for function, name in zip(functions, names):
        y_vals = function(x_vals).astype(float)
        dump['signals'][name] = [
            list(x_vals),
            list(y_vals)
        ]
        plt.plot(x_vals, y_vals, label=name)
        
    plt.grid()
    plt.legend()
    plt.xlim(xmin, xmax)
    plt.ylim(ymin, ymax)
    plt.xlabel('Time [s]')
    plt.ylabel('Voltage [V]')
    
    with open(filename, 'w') as outfile:
        json.dump(dump, outfile)
```

# Tasks

```python
def f1(x):
    return np.sin(x*2*np.pi)

def f2(x):
    return 2*np.sin(x*0.7*2*np.pi)

store_graph([f1, f2], ['CH1', 'CH2'], 'test.json', xmin=-2, xmax=2, ymin=-2.1, ymax=2.1)
```
